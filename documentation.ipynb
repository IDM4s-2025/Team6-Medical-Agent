# Eye Conditions and Symptoms

### Cataracts 
- Blurred or cloudy vision
- Difficulty seeing at night
- Sensitivity to light and glare
- Fading or yellowing of colors
- Double vision in one eye

### Glaucoma 
- Gradual loss of peripheral vision
- Halos around lights
- Eye pain (acute attack)
- Redness
- Nausea/vomiting (acute)
- Headaches

### Age-related Macular Degeneration (AMD) 
- Distorted or blurry central vision
- Difficulty recognizing faces
- Straight lines appear wavy
- Central dark spots

### Diabetic Retinopathy
- Gradual blurring of vision
- Floaters
- Dark areas of vision
- Sudden loss of vision (severe cases)

### Conjunctivitis
- Red or pink eyes
- Itching or burning sensation
- Discharge (watery or thick)
- Eyelid sticking
- Mild photophobia

### Dry Eye Syndrome
- Stinging, burning, or scratchy feeling
- Sensitivity to light
- Eye redness
- Stringy mucus around eyes
- Feeling of "something in the eye"

### Blepharitis
- Sore eyelids
- Itchy eyes
- A gritty feeling in the eyes
- Flakes or crusts around the roots of the eyelashes
- Eyelids sticking together in the morning when you wake up

### Uveitis

- Eye pain, which may feel worse when you're reading
- Red or watery eyes
- Reduced or blurred vision
- Sensitivity to light
- Lights in the corner of your eyes or dark spots (flashes and floaters)

### Retinal Detachment

- Floaters (dots and lines) suddenly appear in your vision or suddenly increase in number
- Flashes of light in your vision
- Dark "curtain" or shadow moving across your vision
- Vision gets suddenly blurred


### Strabismus/Diplopia

- Double vision (diplopia)
- Eyes that do not look in the same direction at the same time
- Eye strain
- Difficulty with depth perception
- Tilting the head to see better

### Stye (Hordeolum)

- Small and painful lump on or inside the eyelid
- Watery eyes
- Eyelid Pain
- Swelling of eyelid
- Eye Redness
- Visible pus in the center of the lump

### Amblyopia
- Squint
- Headaches 
- Tired Eyes
- Difficulty catching or throwing
- Shutting 1 eye or squinting when looking at things
- Blinking a lot
- Not being able to follow an object or person with your eyes
